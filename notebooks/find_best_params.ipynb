```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def extract_csv():
    csv_data = ["k,tau,lambda,gamma,training_rmse,testing_rmse,gap\n"]
    file = open("../recap.txt")
    recap_lines = file.readlines()
    file.close()
    for recap_line in recap_lines:
        cols = recap_line.split(', ')
        line_tab = []
        for col in cols:
            data = col.split("=")[1].strip()
            line_tab.append(data)
        line = ",".join(line_tab) + '\n'
        csv_data.append(line)
    
    file = open("../recap_100k.csv", "a")
    file.writelines(csv_data)
    file.close()
```
